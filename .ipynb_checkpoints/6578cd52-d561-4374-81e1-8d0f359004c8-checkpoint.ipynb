```python
# –ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º phik_matrix
test_features_without_id = test_features.drop(columns=['id'])

# –°–ø–∏—Å–æ–∫ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
categorical_cols = ['dept', 'level', 'workload', 'last_year_promo', 'last_year_violations']
interval_cols = ['employment_years', 'supervisor_evaluation', 'salary']

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –º–∞—Ç—Ä–∏—Ü—ã –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ —Å —è–≤–Ω—ã–º —É–∫–∞–∑–∞–Ω–∏–µ–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
correlation_matrix_test_features = test_features_without_id.phik_matrix(interval_cols=test_features_without_id.columns.difference(categorical_cols))

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –º–∞—Ç—Ä–∏—Ü—ã –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_test_features, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('–ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –≤ test_features')
plt.show()
```

## –ó–∞–¥–∞—á–∞ 1 -  –®–∞–≥ 5. –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

```python
# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ SMAPE
def smape(y_true, y_pred):
    return 100 / len(y_true) * np.sum(2 * np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred)))

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–æ—Å—Ç–æ–π –º–æ–¥–µ–ª–∏ –±–µ–∑ –ø–æ–∏—Å–∫–∞ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
try:
    pipe_final.fit(X_train, y_train)
    y_pred = pipe_final.predict(X_test)
    smape_score = smape(y_test, y_pred)
    print(f'–ú–µ—Ç—Ä–∏–∫–∞ SMAPE –ø—Ä–æ—Å—Ç–æ–π –º–æ–¥–µ–ª–∏: {round(smape_score, 4)}')
except Exception as e:
    print("–ü—Ä–æ–∏–∑–æ—à–ª–∞ –æ—à–∏–±–∫–∞ –≤–æ –≤—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è –ø—Ä–æ—Å—Ç–æ–π –º–æ–¥–µ–ª–∏:")
    print(e)

# –ü–æ–¥–±–æ—Ä –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
param_grid = [
    {
        'models': [LinearRegression()],
        'preprocessor__num__scaler': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    {
        'models': [DecisionTreeRegressor(random_state=100)],
        'models__max_depth': range(5, 15),
        'models__max_features': range(5, 15),
        'preprocessor__num__scaler': [StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]

smape_scorer = make_scorer(smape, greater_is_better=False)

# –ü–æ–∏—Å–∫ –ª—É—á—à–∏—Ö –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ —Å –≤—ã–≤–æ–¥–æ–º –æ—à–∏–±–æ–∫
try:
    grid_search = GridSearchCV(pipe_final, param_grid, n_jobs=-1, cv=5, scoring=smape_scorer)
    grid_search.fit(X_train, y_train)
    best_model = grid_search.best_estimator_
    print('–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å –∏ –µ—ë –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:\n\n', best_model)
    print(f'–ú–µ—Ç—Ä–∏–∫–∞ SMAPE –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –ø–æ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏ –Ω–∞ –æ–±—É—á–∞—é—â–∏—Ö –¥–∞–Ω–Ω—ã—Ö: {round(grid_search.best_score_ * (-1), 4)}')
    print(f'–ú–µ—Ç—Ä–∏–∫–∞ SMAPE –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ: {round(smape(y_test, best_model.predict(X_test)), 4)}')
except Exception as e:
    print("–ü—Ä–æ–∏–∑–æ—à–ª–∞ –æ—à–∏–±–∫–∞ –≤–æ –≤—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–µ–π:")
    print(e)
```

```python
from sklearn.dummy import DummyRegressor
from IPython.display import display

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –Ω–∞ –∞–¥–µ–∫–≤–∞—Ç–Ω–æ—Å—Ç—å
dummy_regr = DummyRegressor(strategy="mean")
dummy_regr.fit(X_train, y_train)
y_pred_dummy = dummy_regr.predict(X_test)
smape_score_dummy = smape(y_test, y_pred_dummy)

display(f'–ú–µ—Ç—Ä–∏–∫–∞ SMAPE –∫–æ–Ω—Å—Ç–∞–Ω—Ç–Ω–æ–π –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ: {round(smape_score_dummy, 4)}')

if best_model:
    if round(smape(y_test, best_model.predict(X_test)), 4) < round(smape_score_dummy, 4):
        display('–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç –∫–æ–Ω—Å—Ç–∞–Ω—Ç–Ω—É—é –º–æ–¥–µ–ª—å –ø–æ –º–µ—Ç—Ä–∏–∫–µ SMAPE –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ.')
    else:
        display('–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç –∫–æ–Ω—Å—Ç–∞–Ω—Ç–Ω—É—é –º–æ–¥–µ–ª—å –ø–æ –º–µ—Ç—Ä–∏–∫–µ SMAPE –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ. –í–æ–∑–º–æ–∂–Ω–æ, —Å—Ç–æ–∏—Ç –ø–µ—Ä–µ—Å–º–æ—Ç—Ä–µ—Ç—å –ø–æ–¥—Ö–æ–¥ –∫ —Å–æ–∑–¥–∞–Ω–∏—é –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ –≤—ã–±–æ—Ä—É –º–æ–¥–µ–ª–∏.')
```

### –ó–∞–¥–∞—á–∞ 2 - –®–∞–≥ 3.1 –ì—Ä–∞—Ñ–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è train_quit

```python
quantitative_columns = ['employment_years', 'supervisor_evaluation', 'salary']

# –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –¥–ª—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
for column in quantitative_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(train_quit[column], kde=True)
    plt.title(f'–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ {column}')
    plt.xlabel(column)
    plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
    plt.show()

# –ö–æ—Ä–æ–±—á–∞—Ç—ã–µ –¥–∏–∞–≥—Ä–∞–º–º—ã –¥–ª—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
for column in quantitative_columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=train_quit[column])
    plt.title(f'–ö–æ—Ä–æ–±—á–∞—Ç–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞ {column}')
    plt.xlabel(column)
    plt.show()

# –ì—Ä–∞—Ñ–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
categorical_columns = ['dept', 'level', 'workload', 'last_year_promo', 'last_year_violations', 'quit']

# –°—Ç–æ–ª–±—á–∞—Ç—ã–µ –¥–∏–∞–≥—Ä–∞–º–º—ã –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=train_quit[column])
    plt.title(f'–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ {column}')
    plt.xlabel(column)
    plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
    plt.show()
```

### –ó–∞–¥–∞—á–∞ 2 - –®–∞–≥ 3.2 –ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º phik_matrix –¥–ª—è train_quit

```python
train_quit_without_id = train_quit.drop(columns=['id'])

# –†–∞–∑–¥–µ–ª—è–µ–º –¥–∞—Ç–∞—Å–µ—Ç –Ω–∞ –¥–≤–µ —á–∞—Å—Ç–∏
quit_data = train_quit_without_id[train_quit_without_id['quit'] == 'yes']
not_quit_data = train_quit_without_id[train_quit_without_id['quit'] == 'no']


# –°—á–∏—Ç–∞–µ–º –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é Phi_k –¥–ª—è —É—à–µ–¥—à–∏—Ö
correlation_matrix_quit = quit_data.phik_matrix(interval_cols=quantitative_columns)

# –°—á–∏—Ç–∞–µ–º –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é Phi_k –¥–ª—è –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è
correlation_matrix_not_quit = not_quit_data.phik_matrix(interval_cols=quantitative_columns)

# –í–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é –¥–ª—è —É—à–µ–¥—à–∏—Ö
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_quit, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('–ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –¥–ª—è —É—à–µ–¥—à–∏—Ö —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤')
plt.show()

# –í–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é –¥–ª—è –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_not_quit, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('–ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –¥–ª—è –Ω–µ—É—à–µ–¥—à–∏—Ö —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤')
plt.show()
```

p.s. –µ—Å–ª–∏ —Ç—ã —Å–ø—Ä–æ—Å–∏—à—å: *—Ä–∞–∑–≤–µ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è –±—É–¥–µ—Ç –ø—Ä–æ—Å—á–∏—Ç—ã–≤–∞—Ç—å—Å—è, –µ—Å–ª–∏ –º—ã –±—É–¥–µ–º –µ–µ —Å—á–∏—Ç–∞—Ç—å –æ—Ç–¥–µ–ª—å–Ω–æ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Å–∞? –ù–∞—Å–∫–æ–ª—å–∫–æ —è –ø–æ–Ω—è–ª- —Ç—ã –ø—Ä–µ–¥–ª–∞–≥–∞–µ—à—å –ø—Ä–æ—Å—á–∏—Ç–∞—Ç—å –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é –æ—Ç–¥–µ–ª—å–Ω–æ –¥–ª—è "–µ–¥–∏–Ω–∏—Ü" –∏ "–Ω—É–ª–µ–π", –Ω–æ –µ—Å–ª–∏ –≤ –∫–æ–ª–æ–Ω–∫–µ –±—É–¥–µ—Ç –≤—Å–µ–≥–æ –æ–¥–Ω–æ –∑–Ω–∞—á–µ–Ω–∏–µ - —Ä–∞–∑–≤–µ –±—É–¥–µ—Ç —Ö–æ—Ç—å —á—Ç–æ-—Ç–æ —Å –Ω–∏–º –∫–æ—Ä—Ä–µ–ª–∏—Ä–æ–≤–∞—Ç—å?*, —Ç–æ –±—É–¥–µ—à—å –∞–±—Å–æ–ª—é—Ç–Ω–æ –ø—Ä–∞–≤: **–µ—Å–ª–∏ –≤ –∫–æ–ª–æ–Ω–∫–µ –µ—Å—Ç—å —Ç–æ–ª—å–∫–æ –æ–¥–Ω–æ –∑–Ω–∞—á–µ–Ω–∏–µ (–Ω–∞–ø—Ä–∏–º–µ—Ä, –≤—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ä–∞–≤–Ω—ã 1 –∏–ª–∏ –≤—Å–µ —Ä–∞–≤–Ω—ã 0) –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞, —Ç–æ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è –±—É–¥–µ—Ç —Ä–∞–≤–Ω–∞ –Ω—É–ª—é. –≠—Ç–æ —Å–≤—è–∑–∞–Ω–æ —Å –æ—Ç—Å—É—Ç—Å—Ç–≤–∏–µ–º –≤–∞—Ä–∏–∞—Ü–∏–∏ –≤ –¥–∞–Ω–Ω—ã—Ö**.

–ù–æ —è –≤—ã—à–µ –Ω–µ –ø—Ä–æ —Ä–∞—Å—á–µ—Ç –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –≥–æ–≤–æ—Ä—é, –∞ –ø—Ä–æ –≤–∏–∑—É–∞–ª—å–Ω—ã–π –∞–Ω–∞–ª–∏–∑ —Ç–æ–≥–æ, –∫–∞–∫–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –æ–∫–∞–∑—ã–≤–∞—é—Ç –≤–ª–∏—è–Ω–∏–µ –Ω–∞ –∫–∞–∂–¥—ã–π –∏–∑ –∫–ª–∞—Å—Å–æ–≤: "—Å–æ—Ç—Ä—É–¥–Ω–∏–∫ —É–≤–æ–ª–∏—Ç—å—Å—è" VS. "—Å–æ—Ç—Ä—É–¥–Ω–∏–∫ –Ω–µ —É–≤–æ–ª–∏—Ç—å—Å—è".

–ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞–≤ –æ—Ç–¥–µ–ª—å–Ω—ã–µ —Ö–∏—Ç–º–∞–ø—ã, —Ç—ã –º–æ–∂–µ—à—å –≤—ã–¥–µ–ª–∏—Ç—å —Ç–µ –ø—Ä–∏–∑–Ω–∞–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ –æ–∫–∞–∑—ã–≤–∞—é—Ç –≤–ª–∏—è–Ω–∏–µ –≤–Ω—É—Ç—Ä–∏ –∫–∞–∂–¥–æ–≥–æ –∏–∑ –∫–ª–∞—Å—Å–æ–≤ —Ç–∞—Ä–≥–µ—Ç–∞. –≠—Ç–æ –º–æ–∂–µ—Ç –ø–æ–º–æ—á—å –≤ –ø–æ–Ω–∏–º–∞–Ω–∏–∏ –≤–∞–∂–Ω–æ—Å—Ç–∏ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã—Ö —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫ –¥–ª—è –ø—Ä–∏–Ω—è—Ç–∏—è —Ä–µ—à–µ–Ω–∏—è –º–æ–¥–µ–ª—å—é.

### –ó–∞–¥–∞—á–∞ 2 - –®–∞–≥ 3.3 –ì—Ä–∞—Ñ–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è test_target_quit

```python
# –ì—Ä–∞—Ñ–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –¥–ª—è test_target_quit
sns.countplot(data=test_target_quit, x='quit')
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ quit')
plt.show()
```

### –ó–∞–¥–∞—á–∞ 2 - –®–∞–≥3.4 –°–æ—Å—Ç–∞–≤—å—Ç–µ –ø–æ—Ä—Ç—Ä–µ—Ç ¬´—É–≤–æ–ª–∏–≤—à–µ–≥–æ—Å—è —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–∞¬ª. –í –∫–∞–∫–æ–º –æ—Ç–¥–µ–ª–µ —Å –±–æ–ª—å—à–µ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é —Ä–∞–±–æ—Ç–∞–µ—Ç —É–≤–æ–ª–∏–≤—à–∏–π—Å—è —Å–æ—Ç—Ä—É–¥–Ω–∏–∫?

```python
# –°–≤–æ–¥–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ –ø–æ –æ—Ç–¥–µ–ª–∞–º
dept_quit = train_quit.groupby(['dept', 'quit']).size().unstack()
dept_quit['quit_rate'] = dept_quit['yes'] / (dept_quit['yes'] + dept_quit['no'])
dept_quit = dept_quit.sort_values(by='quit_rate', ascending=False)

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
plt.figure(figsize=(10, 6))
sns.barplot(x=dept_quit.index, y=dept_quit['quit_rate'])
plt.title('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —É–≤–æ–ª—å–Ω–µ–Ω–∏—è –ø–æ –æ—Ç–¥–µ–ª–∞–º')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞ —É–≤–æ–ª—å–Ω–µ–Ω–∏–π')
plt.xlabel('–û—Ç–¥–µ–ª')
plt.xticks(rotation=45)
plt.show()
```

### –ó–∞–¥–∞—á–∞ 2 - –®–∞–≥3.5 –ö–∞–∫–æ–π —É—Ä–æ–≤–µ–Ω—å –∑–∞–≥—Ä—É–∂–µ–Ω–Ω–æ—Å—Ç–∏ —É —É–≤–æ–ª–∏–≤—à–µ–≥–æ—Å—è —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–∞?

```python
# –°–≤–æ–¥–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ –ø–æ —É—Ä–æ–≤–Ω—é –∑–∞–≥—Ä—É–∂–µ–Ω–Ω–æ—Å—Ç–∏
workload_quit = train_quit.groupby(['workload', 'quit']).size().unstack()
workload_quit['quit_rate'] = workload_quit['yes'] / (workload_quit['yes'] + workload_quit['no'])
workload_quit = workload_quit.sort_values(by='quit_rate', ascending=False)

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
plt.figure(figsize=(10, 6))
sns.barplot(x=workload_quit.index, y=workload_quit['quit_rate'])
plt.title('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —É–≤–æ–ª—å–Ω–µ–Ω–∏—è –ø–æ —É—Ä–æ–≤–Ω—é –∑–∞–≥—Ä—É–∂–µ–Ω–Ω–æ—Å—Ç–∏')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞ —É–≤–æ–ª—å–Ω–µ–Ω–∏–π')
plt.xlabel('–£—Ä–æ–≤–µ–Ω—å –∑–∞–≥—Ä—É–∂–µ–Ω–Ω–æ—Å—Ç–∏')
plt.show()
```

–£—Ä–æ–≤–µ–Ω—å —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç–∏ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –≤—ã—Å–æ–∫–∏–π –≤ –æ—Ç–¥–µ–ª–µ accounting, –Ω–∏–∑–∫–∏–π –≤ sales. –ó–∞–≥—Ä—É–∂–µ–Ω–Ω–æ—Å—Ç—å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ –≤–ª–∏—è–µ—Ç –Ω–∞ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç—å, —Å—Ä–µ–¥–Ω–∏–π —É—Ä–æ–≤–µ–Ω—å –∑–∞–≥—Ä—É–∂–µ–Ω–Ω–æ—Å—Ç–∏ –∞—Å—Å–æ—Ü–∏–∏—Ä—É–µ—Ç—Å—è —Å –Ω–∞–∏–≤—ã—Å—à–µ–π —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç—å—é.

### –ó–∞–¥–∞—á–∞ 2 - –®–∞–≥3.6 –°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–π –∑–∞—Ä–ø–ª–∞—Ç—ã —É–≤–æ–ª–∏–≤—à–∏—Ö—Å—è —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ —Å —Ç–µ–º–∏, –∫—Ç–æ –æ—Å—Ç–∞–ª—Å—è

```python
# –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ test_features –∏ test_target_quit –ø–æ —Å—Ç–æ–ª–±—Ü—É id
merged_test_data = test_features.merge(test_target_quit, on='id')

# –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ train_data –∏ merged_test_data
combined_data = pd.concat([train_data, merged_test_data], ignore_index=True)

# –°–≥—Ä—É–ø–ø–∏—Ä—É–π—Ç–µ –¥–∞–Ω–Ω—ã–µ –ø–æ —É—Ä–æ–≤–Ω—é –¥–æ–ª–∂–Ω–æ—Å—Ç–∏ –∏ —Ä–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ —Å—Ä–µ–¥–Ω—é—é –∑–∞—Ä–ø–ª–∞—Ç—É –¥–ª—è —É—à–µ–¥—à–∏—Ö –∏ –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤
salary_comparison = combined_data.groupby(['level', 'quit'])['salary'].mean().unstack()

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
plt.figure(figsize=(8, 6))
salary_comparison.plot(kind='bar', stacked=True)
plt.title('–°—Ä–µ–¥–Ω—è—è –∑–∞—Ä–ø–ª–∞—Ç–∞ —É–≤–æ–ª–∏–≤—à–∏—Ö—Å—è –∏ –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –ø–æ —É—Ä–æ–≤–Ω—é –¥–æ–ª–∂–Ω–æ—Å—Ç–∏')
plt.ylabel('–°—Ä–µ–¥–Ω—è—è –∑–∞—Ä–ø–ª–∞—Ç–∞')
plt.xlabel('–£—Ä–æ–≤–µ–Ω—å –¥–æ–ª–∂–Ω–æ—Å—Ç–∏')
plt.show()
```

### –ó–∞–¥–∞—á–∞ 2 - –®–∞–≥3.7. –í–ª–∏—è–Ω–∏–µ —É—Ä–æ–≤–Ω—è —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç–∏ —Ä–∞–±–æ—Ç–æ–π –Ω–∞ —É–≤–æ–ª—å–Ω–µ–Ω–∏–µ

```python
test_target = pd.read_csv('https://code.s3.yandex.net/datasets/test_target_job_satisfaction_rate.csv')

# –î–æ–ø—É—Å—Ç–∏–º, —É –Ω–∞—Å –µ—Å—Ç—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —É—Ä–æ–≤–Ω—è —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—ë–Ω–Ω–æ—Å—Ç–∏
predicted_job_satisfaction = test_target.copy()
predicted_job_satisfaction.columns = ['id', 'predicted_job_satisfaction_rate']

# –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –≤—Ç–æ—Ä–æ–π –∑–∞–¥–∞—á–∏ (—É–≤–æ–ª—å–Ω–µ–Ω–∏–µ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤)

# –û–±—ä–µ–¥–∏–Ω—è–µ–º —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –ø–æ id, —É–±–µ–¥–∏–≤—à–∏—Å—å, —á—Ç–æ —Å—Ç–æ–ª–±–µ—Ü 'id' —Å—É—â–µ—Å—Ç–≤—É–µ—Ç
if 'id' in train_data.columns:
    train_quit = pd.merge(train_quit, train_data[['id', 'job_satisfaction_rate']], on='id', how='left')
else:
    display("–û—à–∏–±–∫–∞: –í 'train_data' –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç —Å—Ç–æ–ª–±–µ—Ü 'id'.")

# –û–±—ä–µ–¥–∏–Ω—è–µ–º —Ç–µ—Å—Ç–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ –ø–æ id, —É–±–µ–¥–∏–≤—à–∏—Å—å, —á—Ç–æ —Å—Ç–æ–ª–±–µ—Ü 'id' —Å—É—â–µ—Å—Ç–≤—É–µ—Ç
if 'id' in test_features.columns and 'id' in predicted_job_satisfaction.columns:
    test_features = pd.merge(test_features, predicted_job_satisfaction, on='id', how='left')
else:
    display("–û—à–∏–±–∫–∞: –í 'test_features' –∏–ª–∏ 'predicted_job_satisfaction' –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç —Å—Ç–æ–ª–±–µ—Ü 'id'.")

# –û–±—ä–µ–¥–∏–Ω—è–µ–º —Ç–µ—Å—Ç–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ —Å —Ü–µ–ª–µ–≤—ã–º –ø—Ä–∏–∑–Ω–∞–∫–æ–º —É–≤–æ–ª—å–Ω–µ–Ω–∏—è
test_data_full = pd.merge(test_features, test_target_quit, on='id', how='left')

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —É—Ä–æ–≤–Ω—è —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—ë–Ω–Ω–æ—Å—Ç–∏ –¥–ª—è —É—à–µ–¥—à–∏—Ö –∏ –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤
plt.figure(figsize=(12, 6))
sns.histplot(data=test_data_full, x='predicted_job_satisfaction_rate', hue='quit', multiple='stack', kde=True)
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —É—Ä–æ–≤–Ω—è —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—ë–Ω–Ω–æ—Å—Ç–∏ —Ä–∞–±–æ—Ç–æ–π –¥–ª—è —É—à–µ–¥—à–∏—Ö –∏ –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤')
plt.xlabel('–£—Ä–æ–≤–µ–Ω—å —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—ë–Ω–Ω–æ—Å—Ç–∏ —Ä–∞–±–æ—Ç–æ–π')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤')
plt.show()
```

–°—Ä–µ–¥–Ω—è—è –∑–∞—Ä–ø–ª–∞—Ç–∞ —É–≤–æ–ª—å–Ω—è—é—â–∏—Ö—Å—è —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –Ω–∏–∂–µ, —á–µ–º —É –æ—Å—Ç–∞—é—â–∏—Ö—Å—è. –£—Ä–æ–≤–µ–Ω—å —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç–∏ —Ä–∞–±–æ—Ç–æ–π –¥–ª—è —É–≤–æ–ª—å–Ω—è—é—â–∏—Ö—Å—è –∏ –æ—Å—Ç–∞—é—â–∏—Ö—Å—è —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ —Ä–∞–∑–ª–∏—á–∞–µ—Ç—Å—è: —É —É–≤–æ–ª—å–Ω—è—é—â–∏—Ö—Å—è –æ–Ω –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –Ω–∏–∂–µ (–æ–∫–æ–ª–æ 0.2), –≤ —Ç–æ –≤—Ä–µ–º—è –∫–∞–∫ —É –æ—Å—Ç–∞—é—â–∏—Ö—Å—è - –≤—ã—à–µ 0.6. –≠—Ç–æ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç –≤–ª–∏—è–Ω–∏–µ —É—Ä–æ–≤–Ω—è —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç–∏ –Ω–∞ —É–≤–æ–ª—å–Ω–µ–Ω–∏–µ.

## –ó–∞–¥–∞—á–∞ 2 - –®–∞–≥ 4. –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –Ω–æ–≤–æ–≥–æ –≤—Ö–æ–¥–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞

```python
# –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ test_target –≤ Series, –µ—Å–ª–∏ —ç—Ç–æ DataFrame
if isinstance(test_target, pd.DataFrame):
    test_target = test_target.squeeze()

# –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ test_target –≤ DataFrame, –µ—Å–ª–∏ —ç—Ç–æ Series
if isinstance(test_target, pd.Series):
    test_target = test_target.to_frame(name='job_satisfaction_rate')

# –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö –∏ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞
all_target_features = test_target.merge(test_features, on='id', how='left')
test_data = all_target_features.drop(['job_satisfaction_rate', 'id'], axis=1)
test_target = all_target_features['job_satisfaction_rate']

# –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –æ–±—É—á–∞—é—â–∏–µ –∏ —Ç–µ—Å—Ç–æ–≤—ã–µ
X_train = train_data.drop('job_satisfaction_rate', axis=1)
y_train = train_data['job_satisfaction_rate']
X_test = test_data
y_test = test_target

# –ü–∞–π–ø–ª–∞–π–Ω –¥–ª—è OneHotEncoder –∏ —á–∏—Å–ª–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö
ohe_columns = ['dept', 'last_year_promo', 'last_year_violations', 'level', 'workload']
num_columns = ['employment_years', 'supervisor_evaluation', 'salary']

ohe_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

num_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

data_preprocessor = ColumnTransformer([
    ('ohe', ohe_pipe, ohe_columns),
    ('num', num_pipe, num_columns)
], remainder='passthrough')

pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('model', RandomForestRegressor(n_estimators=100, random_state=42))
])

# –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ –¥–ª—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è job_satisfaction_rate
pipe_final.fit(X_train, y_train)

# –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ —É—Ä–æ–≤–Ω—è —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—ë–Ω–Ω–æ—Å—Ç–∏ –¥–ª—è train_quit –∏ test_features
train_quit['job_satisfaction_rate'] = pipe_final.predict(train_quit[X_train.columns])
test_features['job_satisfaction_rate'] = pipe_final.predict(test_features[X_train.columns])

train_quit_full = pd.concat([train_quit, test_target_quit], axis=0, ignore_index=True)
```

## –ó–∞–¥–∞—á–∞ 2 - –®–∞–≥ 5. –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö

```python
from sklearn.preprocessing import LabelEncoder

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞–ª–∏—á–∏—è —Å—Ç–æ–ª–±—Ü–∞ 'quit' –≤ train_quit_full
if 'quit' in train_quit_full.columns:
    X = train_quit_full.drop(columns=['quit'])
else:
    X = train_quit_full

# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –∫–æ–ª–æ–Ω–∫–∏ 'job_satisfaction_rate' –≤ —á–∏—Å–ª–æ–≤—ã–µ –∫–æ–ª–æ–Ω–∫–∏ –¥–ª—è –≤—Ç–æ—Ä–æ–π –∑–∞–¥–∞—á–∏  
num_columns_with_job_satisfaction = num_columns + ['job_satisfaction_rate']
data_preprocessor_final = ColumnTransformer([
    ('ohe', ohe_pipe, ohe_columns),
    ('num', num_pipe, num_columns_with_job_satisfaction)
], remainder='passthrough')

# –ö–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ —Å –ø–æ–º–æ—â—å—é LabelEncoder
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(train_quit_full['quit'])

# –ü—Ä–µ–ø—Ä–æ—Ü–µ—Å—Å–∏–Ω–≥ –¥–∞–Ω–Ω—ã—Ö
X_train = data_preprocessor_final.fit_transform(X)

# –ò–Ω–≤–µ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ (–µ—Å–ª–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ)
inverted_target = 1 - y_train
print("–ò—Å—Ö–æ–¥–Ω—ã–µ –∫–ª–∞—Å—Å—ã:", y_train)
print("–ò–Ω–≤–µ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –∫–ª–∞—Å—Å—ã:", inverted_target)
```

## –ó–∞–¥–∞—á–∞ 2 - –®–∞–≥ 6. –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏

```python
# –°–æ–∑–¥–∞–Ω–∏–µ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π 'y' –Ω–∞ –æ—Å–Ω–æ–≤–µ —Å—Ç–æ–ª–±—Ü–∞ 'quit'
y = train_quit_full['quit'].apply(lambda x: 1 if x == 'yes' else 0)

# –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –æ–±—É—á–∞—é—â–∏–π –∏ —Ç–µ—Å—Ç–æ–≤—ã–π –Ω–∞–±–æ—Ä—ã
X_train_final, X_test_final, y_train_final, y_test_final = train_test_split(X, y, test_size=0.25, random_state=42)

logistic_model = Pipeline([
    ('preprocessor', data_preprocessor_final),
    ('model', LogisticRegression(max_iter=1000, random_state=42))
])

logistic_model.fit(X_train_final, y_train_final)
logistic_pred = logistic_model.predict_proba(X_test_final)[:, 1]
logistic_roc_auc = roc_auc_score(y_test_final, logistic_pred)
display(f'ROC-AUC for Logistic Regression: {logistic_roc_auc}')

param_grid_rf = {
    'model__n_estimators': [100, 200],
    'model__max_depth': [10, 20, None],
    'model__min_samples_split': [2, 5],
    'model__min_samples_leaf': [1, 2]
}

param_grid_gb = {
    'model__n_estimators': [100, 200],
    'model__learning_rate': [0.1, 0.01],
    'model__max_depth': [3, 5, 7],
    'model__min_samples_split': [2, 5],
    'model__min_samples_leaf': [1, 2]
}

rf_model = Pipeline([
    ('preprocessor', data_preprocessor_final),
    ('model', RandomForestClassifier(random_state=42))
])

grid_search_rf = GridSearchCV(rf_model, param_grid_rf, cv=5, scoring='roc_auc', n_jobs=-1)
grid_search_rf.fit(X_train_final, y_train_final)
rf_best_model = grid_search_rf.best_estimator_
rf_pred = rf_best_model.predict_proba(X_test_final)[:, 1]
rf_roc_auc = roc_auc_score(y_test_final, rf_pred)
display(f'ROC-AUC for Random Forest: {rf_roc_auc}')

gb_model = Pipeline([
    ('preprocessor', data_preprocessor_final),
    ('model', GradientBoostingClassifier(random_state=42))
])

grid_search_gb = GridSearchCV(gb_model, param_grid_gb, cv=5, scoring='roc_auc', n_jobs=-1)
grid_search_gb.fit(X_train_final, y_train_final)
gb_best_model = grid_search_gb.best_estimator_
gb_pred = gb_best_model.predict_proba(X_test_final)[:, 1]
gb_roc_auc = roc_auc_score(y_test_final, gb_pred)
display(f'ROC-AUC for Gradient Boosting: {gb_roc_auc}')

results = {
    'Logistic Regression': logistic_roc_auc,
    'Random Forest': rf_roc_auc,
    'Gradient Boosting': gb_roc_auc
}

best_model_name = max(results, key=results.get)
best_model_score = results[best_model_name]
```

–ü–ª–æ—â–∞–¥—å –ø–æ–¥ ROC-–∫—Ä–∏–≤–æ–π (Area Under Curve ‚Äì –ø–ª–æ—â–∞–¥—å –ø–æ–¥ –∫—Ä–∏–≤–æ–π, Receiver Operating Characteristic ‚Äì —Ä–∞–±–æ—á–∞—è —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∞ –ø—Ä–∏—ë–º–Ω–∏–∫–∞ ... –ø—Ä–∏—à–ª–æ –∏–∑ –æ–±–ª–∞—Å—Ç–∏ —Ä–∞–¥–∏–æ—ç–ª–µ–∫—Ç—Ä–æ–Ω–∏–∫–∏) ‚Äì —ç—Ç–æ –º–µ—Ç—Ä–∏–∫–∞ –æ—Ü–µ–Ω–∫–∏ –¥–ª—è –∑–∞–¥–∞—á –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏. ROC-–∫—Ä–∏–≤–∞—è –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É –¥–æ–ª–µ–π –≤–µ—Ä–Ω–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –æ–±—ä–µ–∫—Ç–æ–≤ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞ –∏ –¥–æ–ª–µ–π –ª–æ–∂–Ω–æ–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –æ–±—ä–µ–∫—Ç–æ–≤. –ù—É –∏ —Å–∞–º–∞ –ø–ª–æ—â–∞–¥—å –ø–æ–¥ ROC-–∫—Ä–∏–≤–æ–π —è–≤–ª—è–µ—Ç—Å—è –º–µ—Ä–æ–π –∫–∞—á–µ—Å—Ç–≤–∞ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞: —á–µ–º –≤—ã—à–µ –∑–Ω–∞—á–µ–Ω–∏–µ AUC (Area Under Curve), —Ç–µ–º –ª—É—á—à–µ –∫–∞—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–π –º–æ–¥–µ–ª–∏. –ù–∞—à –≥–ª–∞–∑ –∏ –º–æ–∑–≥ —É—Å—Ç—Ä–æ–µ–Ω—ã —Ç–∞–∫, —á—Ç–æ –º—ã –æ—á–µ–Ω—å —Ö–æ—Ä–æ—à–æ –≤–∏–¥–∏–º –≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫—É—é —Ä–∞–∑–Ω–∏—Ü—É (–≤ –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ –ø–ª–æ—â–∞–¥–µ–π).
